#### Tratamiento de valores faltantes:

##### Sin tratamiento

```python
import matplotlib.pyplot as plt
import seaborn as sns

def mostrar_distribucion_montos(df, columnas, titulo):
    plt.figure(figsize=(12, 8))
    for i, columna in enumerate(columnas, 1):
        plt.subplot(3, 2, i)
        sns.histplot(df[columna], kde=True, bins=30)
        plt.title(f'Distribución de {columna}')
    plt.suptitle(titulo, fontsize=16)
    plt.tight_layout()
    plt.show()

# Columnas de montos para visualizar
columnas_montos = [
    'monto_compras_nac', 'monto_compras_int',
    'monto_avances_nac', 'monto_avances_int',
    'morosidad', 'pagos_nacional'
]

# Mostrar distribución antes del tratamiento
mostrar_distribucion_montos(df_analisis, columnas_montos, 
                    "Distribución de los datos antes de la imputación")

# Mostrar estadísticas básicas
print("\nEstadísticas descriptivas antes de la imputación:")
print(df_analisis[columnas_montos].describe())
```

##### KNNImputer

```python
from sklearn.impute import KNNImputer

# Inicializar el imputador
imputer_montos = KNNImputer(n_neighbors=5)

# Imputar valores faltantes
df_montos_imputado = pd.DataFrame(
    imputer_montos.fit_transform(df_analisis[columnas_montos]),
    columns=columnas_montos
)

# Mostrar distribución después de imputación
mostrar_distribucion_montos(df_montos_imputado, columnas_montos,
                    "Distribución después de imputación KNN")

# Mostrar estadísticas descriptivas después de la imputación
print("\nEstadísticas descriptivas después de la imputación:")
print(df_montos_imputado.describe())
```

Los gráficos muestran la distribución de monto_compras_nac antes (azul) y después (verde) de aplicar KNNImputer. Se puede observar que la imputación mantuvo la forma general de la distribución mientras completaba los valores faltantes, preservando las características esenciales de los datos originales.

```python
# Crear un DataFrame para las distribuciones a graficar
df_comparacion = pd.DataFrame({
    'Monto Original': df_analisis['monto_compras_nac'],
    'Monto Imputado KNN': df_montos_imputado['monto_compras_nac']
})

# Graficar las distribuciones
plt.figure(figsize=(14, 8))

# Subplot para datos originales
plt.subplot(1, 2, 1)
sns.histplot(df_comparacion['Monto Original'], kde=True, color='blue', label='Original', bins=30)
plt.title('Distribución Original')
plt.xlabel('Monto Compras Nacional')

# Subplot para datos después de KNN
plt.subplot(1, 2, 2)
sns.histplot(df_comparacion['Monto Imputado KNN'], kde=True, color='green', label='Imputada', bins=30)
plt.title('Después de Imputación KNN')
plt.xlabel('Monto Compras Nacional')

# Ajustar el diseño
plt.tight_layout()
plt.show()

# Mostrar estadísticas descriptivas de cada etapa
print("\nEstadísticas descriptivas antes y después de la imputación:")
print(df_comparacion.describe())
```

#### Valores Atípicos

##### Z-Score

```python
from scipy.stats import zscore
import numpy as np

def tratar_atipicos_zscore(df, columnas_numericas, umbral=3):
    df_tratado = df.copy()
    
    for columna in columnas_numericas:
        # Calcular z-scores
        z_scores = zscore(df[columna])
        abs_z_scores = np.abs(z_scores)
        
        # Identificar y reemplazar valores atípicos con la mediana
        mask = abs_z_scores > umbral
        df_tratado.loc[mask, columna] = df[columna].median()
        
        # Mostrar información sobre los cambios
        n_outliers = mask.sum()
        print(f"\nColumna {columna}:")
        print(f"- Número de valores atípicos detectados: {n_outliers}")
        print(f"- Porcentaje de valores atípicos: {(n_outliers/len(df))*100:.2f}%")
    
    return df_tratado

# Aplicar tratamiento de valores atípicos
df_outliers_morosidad = tratar_atipicos_zscore(df_montos_imputado, columnas_montos)

# Mostrar distribución después de tratar outliers
mostrar_distribucion_montos(df_outliers_morosidad, columnas_montos,
                    "Distribución después de tratar outliers con Z-Score")

# Mostrar estadísticas descriptivas después del tratamiento
print("\nEstadísticas descriptivas después del tratamiento de outliers:")
print(df_outliers_morosidad.describe())
```

#### Escalamiento de datos

##### StandardScaler

```python
from sklearn.preprocessing import StandardScaler

# Crear un objeto StandardScaler
scaler = StandardScaler()

# Estandarizar los datos del dataset 
df_outliers_scaled = scaler.fit_transform(df_outliers_morosidad)

# Convertir el resultado en un DataFrame
df_outliers_scaled = pd.DataFrame(df_outliers_scaled, columns=df_outliers_morosidad.columns)

# Mostrar distribución después del escalamiento
mostrar_distribucion_montos(df_outliers_scaled, columnas_montos,
                    "Distribución después del escalamiento StandardScaler")

# Mostrar estadísticas descriptivas después del escalamiento
print("\nEstadísticas descriptivas después del escalamiento:")
print(df_outliers_scaled.describe())
```

Los gráficos muestran la distribución de monto_compras_nac antes (azul) y después (rojo) de aplicar el método Z-Score para el tratamiento de valores atípicos. Se observa cómo el tratamiento suavizó los valores extremos, resultando en una distribución más acotada y manejable para análisis posteriores.

```python
# Crear un DataFrame para las distribuciones a graficar
df_comparacion = pd.DataFrame({
    'Antes de Outliers': df_montos_imputado['monto_compras_nac'],
    'Después de Outliers': df_outliers_morosidad['monto_compras_nac']
})

# Graficar las distribuciones
plt.figure(figsize=(14, 8))

# Subplot para datos antes de tratar outliers
plt.subplot(1, 2, 1)
sns.histplot(df_comparacion['Antes de Outliers'], kde=True, color='blue', label='Original', bins=30)
plt.title('Antes de Tratar Outliers')
plt.xlabel('Monto Compras Nacional')

# Subplot para datos después de tratar outliers
plt.subplot(1, 2, 2)
sns.histplot(df_comparacion['Después de Outliers'], kde=True, color='red', label='Sin Outliers', bins=30)
plt.title('Después de Tratar Outliers')
plt.xlabel('Monto Compras Nacional')

# Ajustar el diseño
plt.tight_layout()
plt.show()

# Mostrar estadísticas descriptivas
print("\nEstadísticas descriptivas antes y después del tratamiento de outliers:")
print(df_comparacion.describe())
```

#### **Valores Atípicos:**

##### **Con IQR** 

El método IQR detecta valores atípicos basándose en la dispersión de los datos dentro del rango intercuartílico, que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Este método es ideal para distribuciones no normales o cuando hay sesgo en los datos, ya que es resistente a valores extremos.

```python
import numpy as np
def tratar_outliers_iqr(df, cols):
    df_tratado = df.copy()
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        # Reemplazar valores fuera de los límites con los límites correspondientes
        df_tratado[col] = np.where(
            df[col] < limite_inferior, 
            limite_inferior, 
            np.where(df[col] > limite_superior, limite_superior, df[col])
        )
    return df_tratado
```

```python
df_outliers_iqr = tratar_outliers_iqr(df_seleccion_knn, df_seleccion_knn.columns)
```

##### **Con Z-Score** 

El método Z-Score identifica valores atípicos calculando cuántas desviaciones estándar (z-scores) se encuentra un dato con respecto a la media de su distribución. Generalmente, los valores con z-scores mayores a 3 o menores a -3 son considerados outliers. Este método es especialmente útil para detectar atípicos en datos con distribuciones normales o simétricas, ya que aprovecha la relación entre la media y la desviación estándar.

```python
from scipy.stats import zscore

def tratar_outliers_zscore(df, cols, umbral=3):
    df_tratado = df.copy()
    for col in cols:
        z_scores = zscore(df[col])
        df_tratado[col] = np.where(
            np.abs(z_scores) > umbral, 
            np.nan,  # Reemplazar valores extremos con NaN
            df[col]
        )
        # Imputar valores NaN con la mediana de la columna
        df_tratado[col] = df_tratado[col].fillna(df[col].median())
    return df_tratado
```

```python
df_outliers_zscore = tratar_outliers_zscore(df_seleccion_knn, df_seleccion_knn.columns)
```

En este caso seleccionamos el método Z-Score para tratar los valores atípicos porque preserva mejor la forma original de la distribución, eliminando únicamente los valores más extremos sin alterar significativamente el rango de los datos restantes. Esto es especialmente útil en situaciones donde los valores extremos no son errores, sino parte de la variabilidad inherente del dataset. Además, el método es adecuado para datos con una distribución cercana a la normal, lo que permite un análisis más representativo sin distorsionar la información clave del conjunto de datos.

```python
# 2. Análisis de valores atípicos con IQR
def detectar_outliers_iqr(df):
    outliers_dict = {}
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        outliers_dict[column] = {
            'total_outliers': len(df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]),
            'porcentaje_outliers': len(df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]) / len(df) * 100
        }
    return pd.DataFrame(outliers_dict).T

print("\nAnálisis de outliers usando IQR:")
print(detectar_outliers_iqr(df_seleccion))
```

Este índice es útil para determinar el número óptimo de clusters y evaluar la efectividad del modelo de agrupamiento.

```python
from sklearn.metrics import silhouette_score

silhouette = silhouette_score(X_sample, y_pred)
print(f"Silhouette Score: {silhouette:.4f}")
```
