#### **Valores Atípicos:**

##### **Con IQR** 

El método IQR detecta valores atípicos basándose en la dispersión de los datos dentro del rango intercuartílico, que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Este método es ideal para distribuciones no normales o cuando hay sesgo en los datos, ya que es resistente a valores extremos.

```python
def tratar_outliers_iqr(df, cols):
    df_tratado = df.copy()
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        # Reemplazar valores fuera de los límites con los límites correspondientes
        df_tratado[col] = np.where(
            df[col] < limite_inferior, 
            limite_inferior, 
            np.where(df[col] > limite_superior, limite_superior, df[col])
        )
    return df_tratado
```

##### **Con Z-Score** 

El método Z-Score identifica valores atípicos calculando cuántas desviaciones estándar (z-scores) se encuentra un dato con respecto a la media de su distribución. Generalmente, los valores con z-scores mayores a 3 o menores a -3 son considerados outliers. Este método es especialmente útil para detectar atípicos en datos con distribuciones normales o simétricas, ya que aprovecha la relación entre la media y la desviación estándar.

```python
def tratar_outliers_zscore(df, cols, umbral=3):
    df_tratado = df.copy()
    for col in cols:
        z_scores = zscore(df[col])
        df_tratado[col] = np.where(
            np.abs(z_scores) > umbral, 
            np.nan,  # Reemplazar valores extremos con NaN
            df[col]
        )
        # Imputar valores NaN con la mediana de la columna
        df_tratado[col] = df_tratado[col].fillna(df[col].median())
    return df_tratado
```

```python
df_outliers_zscore = tratar_outliers_zscore(df_seleccion_knn, df_seleccion_knn.columns)
```

En este caso seleccionamos el método Z-Score para tratar los valores atípicos porque preserva mejor la forma original de la distribución, eliminando únicamente los valores más extremos sin alterar significativamente el rango de los datos restantes. Esto es especialmente útil en situaciones donde los valores extremos no son errores, sino parte de la variabilidad inherente del dataset. Además, el método es adecuado para datos con una distribución cercana a la normal, lo que permite un análisis más representativo sin distorsionar la información clave del conjunto de datos.

#### **Escalamiento de Datos**

##### **StandardScaler**

En este caso, se utiliza **StandardScaler**, que ajusta las características a una media de 0 y una desviación estándar de 1. Esto mejora la eficiencia y estabilidad del modelo, especialmente cuando se usa regularización como en Ridge o Lasso, asegurando que ninguna característica influya desproporcionadamente.
