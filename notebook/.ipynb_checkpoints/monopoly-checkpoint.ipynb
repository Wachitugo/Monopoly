##### **Sin Tratamiento**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Función para mostrar los histogramas
def mostrar_distribucion(df, columnas, titulo):
    plt.figure(figsize=(12, 8))
    for i, columna in enumerate(columnas, 1):
        plt.subplot(3, 5, i)
        sns.histplot(df[columna], kde=True, bins=30)
        plt.title(f'Distribución de {columna}')
    plt.suptitle(titulo, fontsize=16)
    plt.tight_layout()
    plt.show()

# Mostrar distribución de los datos antes de la imputación
mostrar_distribucion(df_categorizado, df_categorizado.columns, "Distribución de los datos antes de la imputación")
```

```python
from sklearn.impute import SimpleImputer

# Crear una copia del DataFrame original
df_seleccion_media = df_categorizado.copy()

# Inicializamos el imputador para todas las columnas 
imputador_media = SimpleImputer(strategy='mean')

# Aplicamos imputación con la media a todas las columnas
df_seleccion_media[:] = imputador_media.fit_transform(df_seleccion_media)

# Mostrar distribución de los datos después de la imputación
mostrar_distribucion(df_seleccion_media, df_seleccion_media.columns, "Distribución de los datos después de la imputación (media)")
```

##### **KNNImputer**

```python
from sklearn.impute import KNNImputer

# Inicializamos el imputador KNN para todas las columnas
knn_imputador = KNNImputer(n_neighbors=5)

# Copia del DataFrame original
df_seleccion_knn = df_categorizado.copy()

# Aplicamos KNNImputer a todas las columnas 
df_seleccion_knn[:] = knn_imputador.fit_transform(df_seleccion_knn)

# Mostrar distribución de los datos después de la imputación
mostrar_distribucion(df_seleccion_knn, df_seleccion_knn.columns, "Distribución de los datos después de la imputación con KNN")
```

#### **Valores Atípicos:**

##### **Con IQR** 

El método IQR detecta valores atípicos basándose en la dispersión de los datos dentro del rango intercuartílico, que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Este método es ideal para distribuciones no normales o cuando hay sesgo en los datos, ya que es resistente a valores extremos.

```python
import numpy as np
def tratar_outliers_iqr(df, cols):
    df_tratado = df.copy()
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        # Reemplazar valores fuera de los límites con los límites correspondientes
        df_tratado[col] = np.where(
            df[col] < limite_inferior, 
            limite_inferior, 
            np.where(df[col] > limite_superior, limite_superior, df[col])
        )
    return df_tratado
```

```python
df_outliers_iqr = tratar_outliers_iqr(df_seleccion_knn, df_seleccion_knn.columns)
```

##### **Con Z-Score** 

El método Z-Score identifica valores atípicos calculando cuántas desviaciones estándar (z-scores) se encuentra un dato con respecto a la media de su distribución. Generalmente, los valores con z-scores mayores a 3 o menores a -3 son considerados outliers. Este método es especialmente útil para detectar atípicos en datos con distribuciones normales o simétricas, ya que aprovecha la relación entre la media y la desviación estándar.

```python
from scipy.stats import zscore

def tratar_outliers_zscore(df, cols, umbral=3):
    df_tratado = df.copy()
    for col in cols:
        z_scores = zscore(df[col])
        df_tratado[col] = np.where(
            np.abs(z_scores) > umbral, 
            np.nan,  # Reemplazar valores extremos con NaN
            df[col]
        )
        # Imputar valores NaN con la mediana de la columna
        df_tratado[col] = df_tratado[col].fillna(df[col].median())
    return df_tratado
```

```python
df_outliers_zscore = tratar_outliers_zscore(df_seleccion_knn, df_seleccion_knn.columns)
```

```python
plt.figure(figsize=(14, 8))
sns.histplot(df_seleccion_knn['renta'], kde=True, color='blue', label='Original', bins=30)
sns.histplot(df_outliers_iqr['renta'], kde=True, color='green', label='Tratado con IQR', bins=30)
sns.histplot(df_outliers_zscore['renta'], kde=True, color='red', label='Tratado con Z-Score', bins=30)

# Añadir títulos y leyenda
plt.title('Comparación de la Distribución de la Renta\nAntes y Después del Tratamiento de Outliers', fontsize=16)
plt.xlabel('Renta', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Métodos de Tratamiento')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

En este caso seleccionamos el método Z-Score para tratar los valores atípicos porque preserva mejor la forma original de la distribución, eliminando únicamente los valores más extremos sin alterar significativamente el rango de los datos restantes. Esto es especialmente útil en situaciones donde los valores extremos no son errores, sino parte de la variabilidad inherente del dataset. Además, el método es adecuado para datos con una distribución cercana a la normal, lo que permite un análisis más representativo sin distorsionar la información clave del conjunto de datos.

### **SVM**

El SVM (Support Vector Machine) es un algoritmo de aprendizaje supervisado utilizado para clasificación y regresión. En el caso de la regresión, conocido como SVR (Support Vector Regression), el objetivo es encontrar un hiperplano en un espacio de alta dimensión que se ajuste lo mejor posible a los datos, minimizando el error de predicción. El modelo trata de mantener un margen pequeño entre los datos y el hiperplano, penalizando más los puntos fuera de este margen. Para obtener buenos resultados, es importante ajustar parámetros como el C (penalización) y el epsilon (margen de error), así como usar un kernel adecuado para transformar los datos en un espacio de mayor dimensión si es necesario.

#### **Separación de Variables** 

El primer paso es separar las variables dependientes e independientes. En este caso, la variable dependiente es la que deseamos predecir, y las variables independientes son aquellas que utilizará el modelo para hacer la predicción.

A continuación, dividimos los datos en conjuntos de entrenamiento y prueba. Esto nos permite entrenar el modelo en un conjunto y evaluar su rendimiento en un conjunto no visto.

#### **Generar el Modelo**

Generamos el modelo SVM utilizando la clase SVR (Support Vector Regression) para problemas de regresión. Podemos seleccionar un kernel (por ejemplo, "linear", "rbf", etc.) y ajustar otros parámetros como C y epsilon para controlar la complejidad y los márgenes de error.

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    svr_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error'
)
```

### **Cluster Jerárquico**

El Cluster Jerárquico es un método de agrupamiento que organiza los datos en una estructura de árbol, permitiendo una visualización clara de las relaciones entre las diferentes observaciones. Este método no requiere que se especifique el número de clusters de antemano y puede ser útil para encontrar patrones ocultos en los datos.

#### **Separación de Variables** 

En este paso, se separan las variables predictoras (o características) de la variable objetivo. En el contexto de un modelo de agrupamiento, generalmente solo se utilizan las variables predictoras. Por ejemplo, si se tiene un conjunto de datos con variables como Edad, Renta, Sexo, entre otras, estas variables se separan de la variable objetivo (si existiera) y se utilizan para realizar el agrupamiento.

```python
X = df_outliers_scaled.drop('uso_total_credito', axis=1)  
y = df_outliers_scaled['uso_total_credito'] 
```

#### **Escalar datos** 

El escalado de datos es crucial para el rendimiento de los algoritmos de agrupamiento, ya que asegura que las variables con mayores magnitudes no dominen el cálculo de las distancias. Utilizamos el StandardScaler para normalizar las características a una escala común, generalmente con media 0 y desviación estándar 1.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 
```

#### **Sample** 

A veces, cuando se trabaja con conjuntos de datos grandes, se puede optar por tomar una muestra representativa de los datos para acelerar el proceso de modelado. En este caso, estamos tomando una muestra de las primeras 10,000 observaciones.

```python
X_sample = X_scaled[:10000] 
```

#### **Cargar el modelo**

Una vez que los datos están listos, se carga el modelo de AgglomerativeClustering, que es el algoritmo de cluster jerárquico utilizado. Este modelo agrupa las observaciones basándose en las distancias entre ellas, y no requiere especificar previamente el número de clusters. Después de ajustar el modelo a los datos, se realiza la predicción de los grupos a los que pertenece cada observación.

```python
from sklearn.cluster import AgglomerativeClustering

agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward') 
y_pred = agg_clustering.fit_predict(X_sample) 
```

### **Cluster Jerárquico**

#### **Silhouette Score** 

El Silhouette Score es una métrica utilizada para evaluar la calidad de un modelo de clustering. Mide qué tan similares son los puntos dentro de un mismo grupo en comparación con los puntos en otros grupos. Su valor oscila entre -1 y 1, donde un valor cercano a 1 indica que los puntos están bien agrupados, un valor cercano a 0 indica que los puntos están en los límites de los grupos, y un valor cercano a -1 sugiere que los puntos están mal agrupados, es decir, podrían pertenecer a otro grupo. Este índice es útil para determinar el número óptimo de clusters y evaluar la efectividad del modelo de agrupamiento.

```python
from sklearn.metrics import silhouette_score

silhouette = silhouette_score(X_sample, y_pred)
print(f"Silhouette Score: {silhouette:.4f}")
```

```python
silhouette = silhouette_score(X_sample, y_pred)
print(f"Silhouette Score: {silhouette:.4f}")

# Visualizamos los clusters generados
plt.scatter(X_sample[:, 0], X_sample[:, 1], c=y_pred, cmap='viridis')
plt.title("Clustering Jerárquico")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar()
plt.show()
```

Un **Silhouette Score de 0.3760** indica que, aunque los grupos formados por el modelo de clustering no están mal definidos, aún hay espacio para mejorar la calidad del agrupamiento. Este valor sugiere que algunos puntos de los clusters pueden estar cerca de los límites de otros clusters, lo que podría implicar que la separación entre grupos no es tan clara como podría ser. Idealmente, un valor más cercano a 1 sería preferible, ya que reflejaría una separación más clara entre los grupos. Sin embargo, un valor de 0.3760 aún puede ser aceptable dependiendo del contexto y los objetivos del análisis.
